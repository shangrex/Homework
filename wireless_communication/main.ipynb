```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from tqdm import tqdm
import random
```

```python
def pos(t, lam):
    return 1-math.exp(-lam*t)
```

```python
def sim(prob):
    poll = np.arange(1, 1001)
    select = np.random.choice(poll)
    return select <= prob*1000
```

```python
arrival_num = 0
block_num = 0
queue_num = 0
server_num = 0
Q = 0
S = 1
lam = 0.01
mu = 0.01
time = 0

time_count = np.zeros(S+1)
leav_prob = np.zeros(S)

departure = np.zeros(100000)
arrival = np.zeros(100000)
blocked = np.zeros(100000)
time_unit = np.arange(100000)

for runtime in range(100000):
    time_count[0] += 1
    for i in range(1, S+1):
        if(time_count[i] > 0):
            time_count[i] += 1
    
    #cal arrival prob.
    arrival_prob = pos(time_count[0], lam)
    
    for i in range(S):
        #cal leaving prob.
        leav_prob[i] = pos(time_count[i+1], mu)
        #check leave
        if(sim(leav_prob[i])):
            time_count[i+1] = 0
            if(queue_num > 0):
                queue_num -= 1
            server_num -= 1  
            #add deaparture
            departure[time] += 1
    
    #check arriavl
    if(sim(arrival_prob)):
        arrival[time] += 1
        arrival_num +=1
        time_count[0] = 0
        #check block
        if(queue_num+server_num > S+Q-1):
            block_num += 1
            blocked[time] += 1
        else:
            if(server_num < S and server_num >= 0):
                server_num += 1
                time_count[server_num] += 1
            elif(queue_num < Q and queue_num >= 0):
                queue_num += 1
            else:
                print("wtf")
    time += 1
    
    
print(arrival_num)
print(block_num)
print(block_num/arrival_num*100, "%")
```
