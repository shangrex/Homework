```python
plt.figure(figsize=(15, 5))
plt.title('No thread Disk R/W')
plt.plot(range(54), nt_data['disk:w'], label='w')
plt.plot(range(54), nt_data['disk:r'], label='r')
plt.legend(bbox_to_anchor=(0.5, 0))
```

```python
plt.figure(figsize=(15, 5))
plt.title('No thread I/O')
plt.plot(range(54), nt_data['io:w'], label='w')
plt.plot(range(54), nt_data['io:r'], label='r')
plt.legend(bbox_to_anchor=(0.5, 0))
plt.show()
```

```python
fig_d = plt.figure(figsize=(140, 10))
plt.title('No thread cpu usage distribution')
plt.plot(range(53), ntc_data['cpu:usr'], label='usr')
plt.plot(range(53), ntc_data['cpu:sys'], label='sys')
plt.plot(range(53), ntc_data['cpu:wai'], label='wai')
plt.plot(range(53), ntc_data['cpu:stl'], label='stl')
plt.legend(bbox_to_anchor=(0.5, 0))
fig_d.show()

fig_c = plt.figure(figsize=(140, 10))
plt.title('No thread cpu core usage')
plt.plot(range(53), ntc_data['cpu0:usr'], label='cpu0')
plt.plot(range(53), ntc_data['cpu1:usr'], label='cpu1')
plt.plot(range(53), ntc_data['cpu2:usr'], label='cpu2')
plt.plot(range(53), ntc_data['cpu3:usr'], label='cpu3')
plt.plot(range(53), ntc_data['cpu4:usr'], label='cpu4')
plt.plot(range(53), ntc_data['cpu5:usr'], label='cpu5')
plt.plot(range(53), ntc_data['cpu6:usr'], label='cpu6')
plt.plot(range(53), ntc_data['cpu7:usr'], label='cpu7')
plt.legend(bbox_to_anchor=(0.5, 0))
fig_c.show()

fig_t = plt.figure(figsize=(140, 10))
plt.title('No thread cpu usage')
plt.plot(range(53), ntc_data['cpu:usr'], label='cpu')
plt.legend(bbox_to_anchor=(0.5, 0))
fig_t.show()
```

```python
plt.figure(figsize=(140, 10))
plt.title('No thread memory usage')
plt.plot(range(54), nt_data['mem:used'], label='used')
plt.plot(range(54), nt_data['mem:free'], label='free')
plt.plot(range(54), nt_data['mem:buff'], label='buff')
plt.plot(range(54), nt_data['mem:cache'], label='cache')
plt.legend(bbox_to_anchor=(0.5, 0))
plt.show()
```

```python
plt.figure(figsize=(140, 10))
plt.title('No thread Interrupt')
plt.plot(range(54), nt_data['int:142'], label='142')
plt.plot(range(54), nt_data['int:143'], label='143')
plt.plot(range(54), nt_data['int:144'], label='144')
plt.legend(bbox_to_anchor=(0.5, 0))
plt.ylim(0, 30)
plt.show()
```

```python
plt.figure(figsize=(140, 10))
plt.title('Single Swap usage')
plt.plot(range(nt_data.shape[0]), nt_data['swap:use'], label='used')
plt.plot(range(nt_data.shape[0]), nt_data['swap:free'], label='free')
plt.legend(bbox_to_anchor=(0.5, 0))
plt.show()
```

```python
plt.figure(figsize=(140, 10))
plt.title('No thread System')
plt.plot(range(ntc_data.shape[0]), ntc_data['sys:int'], label='int')
plt.plot(range(ntc_data.shape[0]), ntc_data['sys:csw'], label='csw')
plt.legend(bbox_to_anchor=(0.5, 0))
plt.show()
```

```python
plt.figure(figsize=(140, 10))
plt.title('No thread Process')
plt.plot(range(nt_data.shape[0]), nt_data['pro:run'], label='run')
plt.plot(range(nt_data.shape[0]), nt_data['pro:new'], label='new')
plt.plot(range(nt_data.shape[0]), nt_data['pro:blk'], label='blk')
plt.ylim(0, 10)
plt.legend(bbox_to_anchor=(0.5, 0))
plt.show()
```
